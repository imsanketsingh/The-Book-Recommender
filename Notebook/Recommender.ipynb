## Recommender System (Collaborative Filtering Based)

```python
books.head(1)
```

```python
users.head(1)
```

```python
ratings.head(1)
```

```python
rating_books.head(1)
```

```python
x = rating_books.groupby('User-ID').count()['Book-Rating']>=200
```

```python
x
```

```python
x[x]
```

```python
usersWithGt200Ratings = x[x].index
```

```python
usersWithGt200Ratings
```

```python
filtered_rating = rating_books[rating_books['User-ID'].isin(usersWithGt200Ratings)]
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
final_ratings_grouped1 = final_ratings.groupby('Book-Title').count()['User-ID'].reset_index()
final_ratings_grouped1 = final_ratings_grouped1.rename(columns={'User-ID': 'NumRatings'})
final_ratings_grouped1
```

```python
final_ratings_grouped2 = final_ratings.groupby('Book-Title')['Book-Rating'].mean().reset_index()
final_ratings_grouped2 = final_ratings_grouped2.rename(columns={'Book-Rating': 'Relative Rating'})
final_ratings_grouped2
```

```python
final_ratings_grouped = final_ratings_grouped2.merge(final_ratings_grouped1, on="Book-Title")
final_ratings_grouped
```

```python
final_ratings.drop_duplicates(['Book-Title'], inplace=True)
```

```python
final_ratings
```

```python
final_ratings_grouped_final = final_ratings_grouped.merge(final_ratings, on="Book-Title")
final_ratings_grouped_final = final_ratings_grouped_final[['Book-Title', 'Relative Rating', 'NumRatings', 'Book-Author', 'Year-Of-Publication', 'Image-URL-M']]
final_ratings_grouped_final
```

```python
pivot_table = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pivot_table
```

```python
pivot_table.fillna(0, inplace=True)
```

```python
pivot_table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
score = cosine_similarity(pivot_table)
```

```python
score.shape
```

```python
def recommend_books(book_name):
    index = np.where(pivot_table.index==book_name)[0][0]
    distances = sorted(list(enumerate(score[index])), key = lambda x:x[1], reverse = True)
    top5similar = distances[1:6]
    names = [[]]
    for i in top5similar:
        currBookName = pivot_table.index[i[0]]
        imgURL = final_ratings_grouped_final[final_ratings_grouped_final['Book-Title'] == currBookName]['Image-URL-M'].values[0]
        author = final_ratings_grouped_final[final_ratings_grouped_final['Book-Title'] == currBookName]['Book-Author'].values[0]
        rating = final_ratings_grouped_final[final_ratings_grouped_final['Book-Title'] == currBookName]['Relative Rating'].values[0]
        numRatings = final_ratings_grouped_final[final_ratings_grouped_final['Book-Title'] == currBookName]['NumRatings'].values[0]
        yearOfPublication = final_ratings_grouped_final[final_ratings_grouped_final['Book-Title'] == currBookName]['Year-Of-Publication'].values[0]
        
        curr_book = [currBookName, imgURL, author, rating, numRatings, yearOfPublication]
        names.append(curr_book)
    return names
```

```python
recommend_books('Animal Farm')
```
